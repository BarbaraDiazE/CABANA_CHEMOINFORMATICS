```python
class ComputeDescriptors:

    def __init__(self, csv_name):
        self.Data = pd.read_csv(csv_name, sep = ",")
        #self.Data = self.Data[600:605]
    def compute_descriptors(self, archivo_salida):
        Data = self.Data
        HBA = list()
        HBD = list()
        RB = list()
        LogP = list()
        TPSA = list()
        MW = list()
        HeavyAtom = list()
        RingCount = list()
        FractionCSP3 = list()
        smiles = list(self.Data["SMILES"])
        sm  = list()
        print(smiles[423])
        for i in smiles:
            sm.append(Chem.MolFromSmiles(i)) 
        for i in sm:
            HBA.append(Descriptors.NumHAcceptors(i))
            HBD.append(Descriptors.NumHDonors(i))
            RB.append(Descriptors.NumRotatableBonds(i))
            LogP.append(Descriptors.MolLogP(i))
            TPSA.append(Descriptors.TPSA(i))
            MW.append(Descriptors.MolWt(i))
        print("calcule descriptores")
        columns = ["SMILES", "HBA", "HBD", "RB", "LogP", "TPSA", "MW"]
        data = [smiles, HBA, HBD, RB, LogP, TPSA, MW]
        data = np.transpose(data, axes=None)
        Database = pd.DataFrame(
            data = data,
            columns = columns)
        len_smiles= len(list(Data["SMILES"]))
        print(Database.head())
        Database["Library"] = ["Biofacquim" for i in range(len_smiles)] 
        Database["Name"] = Data["Name"]
        Database["ID"] = Data["ID"]
        Database["Kingdom"] = Data["Kingdom"]
        Database = Database
        Database.to_csv(str(archivo_salida) + ".csv", sep = ",")
        print("Descriptores Biofacquim 2.0", Database.head())
        return Database
        
```
